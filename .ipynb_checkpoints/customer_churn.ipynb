**CUSTOMER** **SEGMENTATION**


```python
segmentation_data=pd.DataFrame({
    'Tenure Months':X_test['Tenure Months'],
    'Monthly Charges':X_test['Monthly Charges'],
    'Total Charges':X_test['Total Charges'],
    'Churn Value':churn_prob
})
```

Implementation of K means


```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(segmentation_data)
print(scaled_data[:5])
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,16):
  kmeans=KMeans(n_clusters=k,random_state=42)
  kmeans.fit(scaled_data)
  wcss.append(kmeans.inertia_)
  plt.figure(figsize=(8,6))
plt.plot(range(1,16),wcss,marker='o',linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show
```

```python
df["Segment"] = kmeans.fit_predict(
    df[["Tenure Months", "Monthly Charges", "Total Charges"]]
)
```

```python
segmentation_data
```

```python
kmeans=KMeans(n_clusters=3,random_state=42)
segmentation_data['cluster']=kmeans.fit_predict(scaled_data)
```

```python
clusters=kmeans.fit_predict(scaled_data)
```

```python
segmentation_data['cluster']=clusters
```

```python
cluster_summary=segmentation_data.groupby('cluster').mean()
```

```python
cluster_summary
```

```python
cluster_names={
    0:'Budget Loyal Customers',
    1:'High Risk New customers',
    2:'Loyal Premiuim Customers'
}
```

```python
segmentation_data['Cluster Segment']=segmentation_data['cluster'].map(cluster_names)
```
